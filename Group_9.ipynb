```python
import pandas as pd
heart = pd.read_csv('/content/heart.csv')
heart
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

###Gender vs Heart disease


Target represents presence or absence heart disease
0 = No heart disease
1 = Presence of heart disease

For sex:
0 represents female
1 represents male

```python
sns.countplot(heart,x='sex',hue='target')
plt.title('Heart Disease')
plt.show()
plt.figure(figsize=(16,8))
```

###Relationship among Maximum Heart rate , age and Heart disease

Thalach represents the maximum heart rate an individual has.


```python
sns.scatterplot(heart,x='age',y='thalach',hue='target')
plt.title('Age vs Maximum Heart Rate and Heart Diseases')
plt.show()
```


* There is a negative correlation between maximum heart rate and age
* The higher the maximum heart rate,the greater the risk of heart disease
* The age group that is at greatest risk of getting a heart disease is 40-65





###Age Distribution of Patients

Here we are examining the age distibution and check if age influence the chances of occurence of heart disease.

```python
sns.histplot(heart,x='age',hue='target',kde=1)
plt.title('A histogram showing heart disease by age distribution')
plt.show()
```

###Relationship among cholestral and resting blood pressure(trestbps) against heart disease

```python
sns.lineplot(heart,x='chol',y='trestbps',hue='target')
```

###The distribution of patients with exercise induced angina and oldpeak by heart disease

* Exercise induced angina is chest pain caused by exercise
* Oldpeak is the depression induced by exercice(High values suggest more significant cardiac stress


```python
sns.boxplot(heart,x = 'exang',y = 'oldpeak',hue = 'target')
```


* The median of old peak of healthy patients is higher compared to those with heart disease  
* The distribution of the unhealthy patients is more skewed
* The old peak of the unhealthy is low
* Unhealthy patients without chest pain have the most outliers

```python
sns.violinplot(heart,x = 'exang',y = 'oldpeak',hue = 'target')
```

* Majority of the population is healthy (bigger violins)

```python
sns.boxplot(heart,x = 'exang',y = 'oldpeak',hue = 'target',showfliers = 0)
```
